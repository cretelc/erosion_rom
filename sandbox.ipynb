```python

import numpy as np
import matplotlib.pyplot as plt
import imageio

# Define parameters
num_frames = 20  # Number of frames in the GIF
filenames = []

# Generate plots and save as images
for i in range(num_frames):
    x = np.linspace(0, 2 * np.pi, 100)
    y = np.sin(x + (i / num_frames) * 2 * np.pi)  # Animate the sine wave

    plt.figure(figsize=(5, 3))
    plt.plot(x, y, label=f"Frame {i}")
    plt.legend()
    plt.ylim(-1.2, 1.2)
    
    filename = f"frame_{i}.png"
    plt.savefig(filename)  # Save frame as an image
    filenames.append(filename)
    plt.close()  # Close the figure to avoid memory issues

# Create GIF using imageio
gif_filename = "animation.gif"
with imageio.get_writer(gif_filename, mode="I", duration=0.1) as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)

print(f"GIF saved as {gif_filename}")

```
