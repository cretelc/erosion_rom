```python
import numpy as np
from matplotlib.patches import Polygon
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt

fpath = "grid_geometry.csv"
data = np.genfromtxt(fpath, delimiter=',', skip_header=1)
```

```python
t_interp = np.linspace(0, 20, 21)
print(t_interp)
reduced_ra_ds = np.interp(t_interp, data[:,0], data[:,1])
reduced_ra_us = np.interp(t_interp, data[:,0], data[:,2])
reduced_ta = np.interp(t_interp, data[:,0], data[:,3])
reduced_ts = np.interp(t_interp, data[:,0], data[:,4])
```

```python
rs = 1.9/2
lg = 0.34
for i in range(len(t_interp)):
    ra1 = reduced_ra_ds[i]
    ra2 = reduced_ra_us[i]
    ta = reduced_ta[i]
    ts = reduced_ts[i]
    
    pts = [[lg,ra2], [lg+ta, ra1], [lg+ta, 1.2], [lg,1.2]]
    fig, ax = plt.subplots()
    ax.plot([0,1.5], [0,1.2], alpha=0)
    ax.add_patch(Rectangle((0, rs), -ts, 1.2-rs, edgecolor='red', facecolor='none', lw=2))
    # accel grid
    ax.add_patch(Polygon(pts, edgecolor='red', facecolor='none', lw=2))
    ax.plot(1, 1.2, 'r*')
    ax.grid(which='both')
    ax.set_xlim([-0.5, 1.5])

```
