```python
import numpy as np
import os
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt



fpath = "grid_geometry.csv"
data = np.genfromtxt(fpath, delimiter=',', skip_header=1)
```

```python

t_interp = np.linspace(0, 20, 21)
print(t_interp)
reduced_ra = np.interp(t_interp, data[:,0], data[:,1])
reduced_ta = np.interp(t_interp, data[:,0], data[:,2])
```

```python
for i in range(len(t_interp)):
    ra = reduced_ra[i]
    ta = reduced_ta[i]
    fig, ax = plt.subplots()
    ax.plot([0,1.5], [0,1.2], alpha=0)
    ax.add_patch(Rectangle((1, ra), ta, 1.2-ra, edgecolor='red', facecolor='none', lw=2))
    ax.plot(1, 1.2, 'r*')
    ax.grid(which='both')
```
